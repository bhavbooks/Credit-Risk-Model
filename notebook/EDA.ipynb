#### Rule 1:
All the continuos variable must be numerical

```python
## Handling emp_length column
df['emp_length'].value_counts()

## To numerical
## df['emp_length'] = df['emp_length_int']
```

```python
df['emp_length_int'] = df['emp_length'].str.replace(" years","").str.replace(" year","").str.replace("<1",str(0)).str.replace("10+",str(10))

## checking the conversion
df['emp_length_int'].value_counts()

# Dtype checking, should be str now
type(df['emp_length_int'][0])  # good to go
```
